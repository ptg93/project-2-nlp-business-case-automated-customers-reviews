```python
# Function to determine negative, neutral and positive
def review_sentiment(rating):
    if rating in [1, 2]:
        return "negative"
    elif rating in [3]:
        return "neutral"
    elif rating in [4, 5]:
        return "positive"
    else:
        return None
```

```python
# Apply the reviews rating
data['sentiment'] = data['reviews.rating'].apply(review_sentiment)
data.head()
```

```python


# Function to count words in a text and return True if count > 512
def word_count_above_threshold(text):
    return len(str(text).split()) > 512

# Apply the function to each entry in the "reviews.text" column and create a boolean mask
word_count_mask = data["reviews.text"].apply(word_count_above_threshold)

# Filter new_data to get rows where word count > 512 and select "reviews.text" and their corresponding indexes
reviews_above_threshold = data[word_count_mask][["reviews.text"]]

# Get the indexes of the filtered rows
indexes_above_threshold = reviews_above_threshold.index

# Print the reviews and their corresponding index numbers
for index, review in zip(indexes_above_threshold, reviews_above_threshold["reviews.text"]):
    print(f"Index: {index}, Review: {review}")
```

```python
# import pandas as pd
import nltk
import torch
from transformers import pipeline, AutoTokenizer
from nltk.tokenize import sent_tokenize



# Download NLTK tokenizer data
nltk.download('punkt')

# # Define a function to count words in a text and return True if count > 512
# def word_count_above_threshold(text):
#     return len(str(text).split()) > 512

# # Apply the function to each entry in the "reviews.text" column and create a boolean mask
# word_count_mask = data["reviews.text"].apply(word_count_above_threshold)

# # Filter out rows where the word count is above 512
# filtered_data = data[~word_count_mask]

# Load the tokenizer for the model to handle token length accurately
tokenizer = AutoTokenizer.from_pretrained("sshleifer/distilbart-cnn-12-6")

# Assuming 'data' is already prepared with necessary columns
data = data[["primaryCategories", "brand", "reviews.text", "reviews.rating", "year", "sentiment"]]

# Group and concatenate reviews by 'primaryCategories' and 'sentiment'
grouped_data = data.groupby(['primaryCategories', 'reviews.rating'])['reviews.text'].apply(lambda x: ' '.join(x)).reset_index()

# Load summarizer pipeline
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6", device=0 if torch.cuda.is_available() else -1)

# Function to split text into chunks that fit the model's input size constraints
def split_into_chunks(text, max_tokens=1024):
    tokens = tokenizer(text, truncation=True, max_length=max_tokens, return_tensors="pt")
    input_ids = tokens.input_ids.squeeze(0)
    chunks = []
    for i in range(0, input_ids.size(0), max_tokens):
        chunk_tokens = input_ids[i:i + max_tokens]
        if len(chunk_tokens) > 0:
            chunk_text = tokenizer.decode(chunk_tokens, skip_special_tokens=True)
            chunks.append(chunk_text)
    return chunks

# Function to segment text into sentences
def segment_into_sentences(text):
    sentences = sent_tokenize(text)
    return sentences

# Summarize the grouped reviews
summaries = []

for index, row in grouped_data.iterrows():
    concatenated_reviews = row['reviews.text']
    sentences = segment_into_sentences(concatenated_reviews)
    summary_texts = []
    for sentence in sentences:
        # Avoid summarizing empty sentences
        if sentence.strip():
            summary = summarizer(sentence, max_length=150, min_length=30, do_sample=False)[0]['summary_text']
            summary_texts.append(summary)
    combined_summary = ' '.join(summary_texts)
    summaries.append(combined_summary)

grouped_data['review_summary'] = summaries

# # Summarize the grouped reviews
# summaries = []

# for index, row in grouped_data.iterrows():
#     concatenated_reviews = row['reviews.text']
#     chunks = split_into_chunks(concatenated_reviews)
#     summary_texts = []
#     for chunk in chunks:
#         # Avoid summarizing empty chunks
#         if chunk.strip():
#             summary = summarizer(chunk, max_length=100, min_length=30, do_sample=False)[0]['summary_text']
#             summary_texts.append(summary)
#     combined_summary = ' '.join(summary_texts)
#     summaries.append(combined_summary)

# grouped_data['review_summary'] = summaries

# Save the dataframe to a CSV file
grouped_data.to_csv('summarized_grouped_data.csv', index=False)
```
